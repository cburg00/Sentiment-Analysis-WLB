```python
import matplotlib.pyplot as plt
```

```python
sex_col = df['SOCIO- DEMOGRAPHIC ']['SEX']
sex_counts = sex_col.value_counts()
labels = ('Male', 'Female')

pie = plt.pie(sex_counts, autopct='%1.1f%%')
plt.legend(pie[0], labels, title='Employee Sex')
```

```python
age_col = df['SOCIO- DEMOGRAPHIC ']['AGE']
age_counts = age_col.value_counts()

labels = ('<= 28', '28-38', '38-48', '48-58', '>58')

pie = plt.bar(labels, age_counts, color=['red','blue','yellow','green','purple'])
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Employee Age Groups')
```

```python
satis_col = df['EMPLOYMENT STATUS ']['SATISFACTION']
satis_counts = satis_col.value_counts()

labels = ('Yes','No')

pie = plt.bar(labels, satis_counts, color=['blue', 'red'])
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')
plt.title('Employee Job Satisfaction Distribution')
```
